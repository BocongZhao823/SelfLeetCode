# 122. Best Time to Buy and Sell Stock II

You are given an integer array prices where prices[i] is the price of a given stock on the ith day.

On each day, you may decide to buy and/or sell the stock. You can only hold at most one share of the stock at any time. However, you can buy it then immediately sell it on the same day.

Find and return the maximum profit you can achieve.

### Example 1:

Input: prices = [7,1,5,3,6,4]

Output: 7

Explanation: Buy on day 2 (price = 1) and sell on day 3 (price = 5), profit = 5-1 = 4.

Then buy on day 4 (price = 3) and sell on day 5 (price = 6), profit = 6-3 = 3.

Total profit is 4 + 3 = 7.

### Example 2:

Input: prices = [1,2,3,4,5]

Output: 4

Explanation: Buy on day 1 (price = 1) and sell on day 5 (price = 5), profit = 5-1 = 4.

Total profit is 4.

### Example 3:

Input: prices = [7,6,4,3,1]

Output: 0

Explanation: There is no way to make a positive profit, so we never buy the stock to achieve the maximum profit of 0.
 
### Constraints:

1 <= prices.length <= 3 * 104

0 <= prices[i] <= 104

### My Thinking
1. If I haven't buy any stock at the very begining , I will initial my buy day at Day 0. Then compare the price at Day 1 & Day 0. If price at Day 1 is lower, I won't buy at day 0. As I know if I buy at day 1, I can make more profit. Then compare day 1 & day 2 price, if day 2 price is lower, I won't buy at day 1. Etc... If all days are cheaper than previous day, it means there is no profit to make at all. Profit is 0. O/W, if price at Day 1 is cheaper than Day 2, I will try the case of buying at Day 1 and see what's might going on. 
2. If I already hold a stock, I won't sell it at a lower price (i.e., when buyPrice > sellPrice), as I will make negative profit.
3. I cannot sell in a previous day.
4. It is not worth to buy at the last day, as there is no more days to sell it. And it is not worth to sell at the last day again as will make 0 profit.
5. If I just buy at Day X, it is not worth to sell at Day X, as it will make 0 profit.
6. If I just sell at Day X, it might worth to buy at Day X, as it might make profit in the following days.

```python
class Solution:

    # Find when shall I first buy the stock 
    def buyStock(prices):
        holdStock = False
        n = len(prices)
        buy = prices[0]
        buyDay = -1
        for i in range(n-1): # Won't buy at the last day
            if holdStock == False:
                if buy > prices[i+1]:
                    buy = prices[i+1]
                    buyDay = i + 1
                    #print(buy)
                else:
                    holdStock = True
        return (buyDay, buy, holdStock)

    def maxProfit(prices: list[int]) -> int:
        holdStock = False
        profitList = []
 
        buyDay = Solution.buyStock(prices = prices)[0]
        buy = Solution.buyStock(prices = prices)[1]
        holdStock = Solution.buyStock(prices = prices)[2]
        # print(buyDay)
        # Only care about the prices after I first buy the stock. Not worth to buy & then sell again at the same day
        
        if buyDay > -1:
            prices = prices[buyDay:]
        else:
            return 0 # Won't buy the stock as price is decasing all the time & Won't make any profit at all
        
        # Decide when shall I sell the stock & buy the stock again
        m = len(prices)
        for t in range(m-1): # Won't buy at the last day but can sell at the last day
            print(f"My prices are {prices}")
            print(f"My index is {t}")
            if holdStock == True:
                profit = 0
                if buy < prices[t]:
                    print(f"My current buy price is {buy}")
                    sell = prices[t]
                    print(f"My current sell price is {sell}")
                    profit = sell - buy 
                    profitList.append(profit)
                    print(f"My current profitList {profitList}")
                    prices = prices[t-1:]
                    holdStock = False
                    while len(prices)!= 1 and holdStock == False: # While I did not reach the last day
                        buyDay = Solution.buyStock(prices = prices)[0]
                        buy = Solution.buyStock(prices = prices)[1]
                        holdStock = Solution.buyStock(prices = prices)[2]
                        print("Reach here!")
                        # Buy stock again
        elementProfit = max(profitList)
                
        
        print(elementProfit)
        
        #return(profit)

```
